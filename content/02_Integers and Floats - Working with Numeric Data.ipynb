# Experesion and variables

```python
# Integer (Bilangan Bulat): 
# Mendefinisikan variabel dengan nilai integer
angka_bulat = 5

# Operasi matematika dengan integer
hasil_penjumlahan = angka_bulat + 3
hasil_pengurangan = angka_bulat - 2
hasil_perkalian = angka_bulat * 4
hasil_pembagian = angka_bulat / 2

# Menampilkan hasil
print("Angka bulat:", angka_bulat)
print("Hasil Penjumlahan:", hasil_penjumlahan)
print("Hasil Pengurangan:", hasil_pengurangan)
print("Hasil Perkalian:", hasil_perkalian)
print("Hasil Pembagian:", hasil_pembagian)

# Float (Bilangan Pecahan):
# Mendefinisikan variabel dengan nilai float
angka_pecahan = 3.14

# Operasi matematika dengan float
hasil_penjumlahan_float = angka_pecahan + 1.5
hasil_pengurangan_float = angka_pecahan - 0.5
hasil_perkalian_float = angka_pecahan * 2
hasil_pembagian_float = angka_pecahan / 2

# Menampilkan hasil
print("Angka pecahan:", angka_pecahan)
print("Hasil Penjumlahan (float):", hasil_penjumlahan_float)
print("Hasil Pengurangan (float):", hasil_pengurangan_float)
print("Hasil Perkalian (float):", hasil_perkalian_float)
print("Hasil Pembagian (float):", hasil_pembagian_float)


# Addition
x = 56+65+89+45+78.5+98.2
print(x)
print(type(x))

# Division
x = 125/24
print(x)
print(type(x))

# Floor division
x = 125//24
print(x)
print(type(x))

# Modulus
x = 125%24
print(x)
print(type(x))

# Exponentiation
x = 2**3
print(x)
print(type(x))

# An example: Let's calculate how many minutes there are in 20 hours?
one_hour = 60 # 60 minutes
hour = 20
minutes = one_hour *hour
print(minutes)
print(type(minutes))

# An example: Let's calculate how many hours there are in 348 minutes?
minutes = 348
one_hour = 60
hours = 348/60
print(hours)
print(type(hours))


# Contoh penggunaan fungsi matematika
nilai_absolut = abs(-7)
hasil_pangkat = pow(2, 3)  # 2 pangkat 3
nilai_bulat = round(3.75)

print("Nilai Absolut:", nilai_absolut)
print("Hasil Pangkat:", hasil_pangkat)
print("Nilai Bulat:", nilai_bulat)


#Conversion:
int_to_float = float(5)
float_to_int = int(3.14)

print("int to float:", float_to_int)
print("float to int:", float_to_int)

#Precision and Rounding
rounded_result = round(3.141592653589793, 2)  # Rounds to 2 decimal places

print("rounded result:", rounded_result)
```
